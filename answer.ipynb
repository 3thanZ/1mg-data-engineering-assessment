Use SQL to select all existing columns and adds these new columns:
* One-Hot encoding for each hour of the day /!\ ASSUMPTION: PICKUP DATE
* One-Hot encoding for each day	of the week /!\ ASSUMPTION: PICKUP DATE
* Duration in seconds of the trip
* An int encoding to indicate if the pickup or dropoff locations were at JFK airport. 

```python
q = """
SELECT
    *,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "0" THEN 1 ELSE 0 END AS Weekday_sunday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "1" THEN 1 ELSE 0 END AS Weekday_monday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "2" THEN 1 ELSE 0 END AS Weekday_tuesday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "3" THEN 1 ELSE 0 END AS Weekday_wednesday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "4" THEN 1 ELSE 0 END AS Weekday_thursday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "5" THEN 1 ELSE 0 END AS Weekday_friday,
    CASE WHEN strftime("%w", lpep_pickup_datetime) = "6" THEN 1 ELSE 0 END AS Weekday_saturday,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "0" THEN 1 ELSE 0 END AS Hourofday_0,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "1" THEN 1 ELSE 0 END AS Hourofday_1,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "2" THEN 1 ELSE 0 END AS Hourofday_2,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "3" THEN 1 ELSE 0 END AS Hourofday_3,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "4" THEN 1 ELSE 0 END AS Hourofday_4,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "5" THEN 1 ELSE 0 END AS Hourofday_5,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "6" THEN 1 ELSE 0 END AS Hourofday_6,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "7" THEN 1 ELSE 0 END AS Hourofday_7,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "8" THEN 1 ELSE 0 END AS Hourofday_8,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "9" THEN 1 ELSE 0 END AS Hourofday_9,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "10" THEN 1 ELSE 0 END AS Hourofday_10,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "11" THEN 1 ELSE 0 END AS Hourofday_11,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "12" THEN 1 ELSE 0 END AS Hourofday_12,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "13" THEN 1 ELSE 0 END AS Hourofday_13,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "14" THEN 1 ELSE 0 END AS Hourofday_14,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "15" THEN 1 ELSE 0 END AS Hourofday_15,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "16" THEN 1 ELSE 0 END AS Hourofday_16,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "17" THEN 1 ELSE 0 END AS Hourofday_17,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "18" THEN 1 ELSE 0 END AS Hourofday_18,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "19" THEN 1 ELSE 0 END AS Hourofday_19,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "20" THEN 1 ELSE 0 END AS Hourofday_20,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "21" THEN 1 ELSE 0 END AS Hourofday_21,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "22" THEN 1 ELSE 0 END AS Hourofday_22,
    CASE WHEN strftime("%-H", lpep_pickup_datetime) = "23" THEN 1 ELSE 0 END AS Hourofday_23,
    strftime("%s",Lpep_dropoff_datetime) - strftime("%s",lpep_pickup_datetime) AS Trip_duration_in_seconds,
    CASE WHEN Dropoff_latitude >= 40.641 AND Dropoff_latitude <= 40.803 AND Dropoff_longitude >= -73.933 AND Dropoff_longitude <= -73.792 THEN 1 ELSE 0 END AS Dropoff_in_jfk
FROM trips
"""
ddf = pandas.read_sql(q, con=conn, parse_dates=['lpep_pickup_datetime','Lpep_dropoff_datetime'])
ddf.head()
```
